```python
import ebooklib
from ebooklib import epub
from reportlab.pdfgen import canvas
from PIL import Image
import os, io
```

```python
def epb2pdf(ePath, pPath):
    book = epub.read_epub(ePath)
    # page_size = ((pSize[0]/25.4)*72, (pSize[1]/25.4)*72)
    cvs = canvas.Canvas(pPath)

    for item in book.get_items():
        if isinstance(item, epub.EpubImage):                
            img_data = item.get_content()
            img = Image.open(io.BytesIO(img_data))
            cvs.setPageSize((img.width, img.height))
            cvs.drawInlineImage(img, 0, 0, width=img.width, height=img.height)
    
    cvs.save()
```

```python
size = (235.22, 298.22)
bPath = "/workspace/Books/"
files = os.listdir(bPath)
books = []
for f in files:
    if ".epub" in f:
        books.append(f)
```

```python
for book in books:
    p = bPath + book
    base = book.split(".epub")[0] + ".pdf"
    out = f"../Books/processed/{base}"
    epb2pdf(p, out)
```
