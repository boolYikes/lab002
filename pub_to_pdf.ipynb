```python
# GET BOOK LIST
size = (235.22, 298.22)
bPath = "/workspace/Books/"
files = os.listdir(bPath)
books = []
for f in files:
    if ".epub" in f:
        books.append(f)
```

```python
if not os.path.isdir("./temp"): os.makedirs("./temp")
if not os.path.isdir("./img"): os.makedirs("./img")
imgkit_options = {
        'format': 'png',
        'xvfb': ''
        }
config = imgkit.config(wkhtmltoimage='/usr/bin/wkhtmltoimage', xvfb='/usr/bin/xvfb-run')
for book in books:
    p = bPath + book
    bk = epub.read_epub(p)
    items = bk.get_items_of_type(ebooklib.ITEM_DOCUMENT)
    htmls = []
    for item in items:
        print(item.get_name())
        temp_html = f"./temp/temp_{item.get_name().split('.xhtml')[0]}.html"
        htmls.append(temp_html)
        with open(temp_html, 'wb') as html_file:
            html_file.write(item.content)
    
    for file in htmls:
        img_path = f"./img/img_{item.get_name().split('.xhtml')[0]}.png"
        
        imgkit.from_file(temp_html, img_path, options=imgkit_options, config=config)
        try:
            os.remove(file)
        except Exception as e:
            print(f"os.remove() error: {e}")
            # print(item.get_content())
        
    break
```
